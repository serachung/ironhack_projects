## Preparing full dataset: last 5 years

Now we know the relevance of the metropolitan area of São Paulo, and the great potential there is in this zone (close to main São Paulo city), the analysis will be made focusing mainly in the 39 cities of this area, and in a time frame of 5 years. 

The data was collected in the ANS official site, and will be manipulated in order to optimize memory space, without losing the details.

### Gathering datasets

```python
# Gathering all the information of 5 years in one set:

directory = 'data/'
df_list = [] 

for file in os.listdir(directory):
    if file.endswith("SP.csv"):    
        df = pd.read_csv(f'{directory}{file}', sep=';', encoding = "ISO-8859-1")
        
        for city in metropole_upper:
            city = format_city(city)
            mask = (df['NM_MUNICIPIO'] == city)
            city_df = df.loc[mask,:]
            df_list.append(city_df)
    else:
        continue
        
metrop_raw_df = pd.concat(df_list)
```

```python
metrop_raw_df.head()
```

```python
metrop_raw_df.shape
```

```python
# Saving metropolitan raw df in file:

metrop_raw_df.to_csv('data/treated_data/benef_metropolitana_raw_2015_2020_consolidado.csv')
```

# Side Functions

```python
def show_values_on_bars(axs, num_format, display_units):
    def _show_on_single_plot(ax):        
        for p in ax.patches:
            _x = p.get_x() + p.get_width() / 2
            _y = p.get_y() + p.get_height()
            value = num_format.format((p.get_height()/display_units))
            ax.text(_x, _y, value, ha="center") 

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)
```

```python
def show_values_on_bars_percentage(axs, num_format, display_units):
    def _show_on_single_plot(ax):        
        for p in ax.patches:
            _x = p.get_x() + p.get_width() / 2
            _y = p.get_y() + p.get_height()-1000
            value = num_format.format((p.get_height()/total)*display_units)
            ax.text(_x, _y, value, ha="center") 

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)
```

```python
def format_city (city:str):
    '''
    this function transforms city names in a standardized pattern:
    no accents and all upper case
    '''
    city_clean = city.upper().replace('-',' ').replace('\n',' ').replace ('Á', 'A').replace ('Â', 'A').replace ('Ã', 'A').replace ('É', 'E').replace ('Ê', 'E').replace ('Í', 'I').replace ('Ó', 'O').replace ('Õ', 'O').replace ('Ú', 'U').replace ('Ç', 'C')
    
    if city_clean == 'EMBU':
        city_clean = 'EMBU DAS ARTES'
    
    return city_clean
```

```python
def age_range (age_range:str)-> str:
    
    '''
    This function transforms a text age range string into a '\d+ - \d+' format.
    Example:
        input: '9 a 23 anos'
        output -> '9 - 23'
        
        input: '59 ou mais'
        output: '59 - 999'
    '''
    
    pattern1 = '\d{2}.*\d{2}'
    pattern2 = '(\d{2}).*'    
    
    if len(re.findall(pattern1, age_range)) > 0:
        return re.findall(pattern1, age_range)[0].replace('a','-')
    
    elif len(re.findall(pattern2, age_range)) > 0:
        return re.findall(pattern2, age_range)[0] + ' - ' + '999'
    
    else:
        pass
```
