<b>Questão 1)</b> Considerando os dados do arquivo agrupamento_Q1.csv, execute o algoritmo K-means com 10 iterações tendo como posiçãoo inicial dos centroides de cada cluster as K primeiras linhas do arquivo agrup_centroides_Q1.csv. Por exemplo, para executar o algoritmo com 3 clusters considere os centroides determinados pelas 3 primeiras linhas do arquivo.

Qual a quantidade ideal de clusters? Qual o posicionamento final dos centroides do caso ideal?

```python
data = pd.read_csv('agrupamento_Q1.csv')
centr = pd.read_csv('agrup_centroides_Q1.csv')
```

```python
data.head()
```

```python
centr.head()
```

```python
centr = centr.drop('Unnamed: 0', axis=1)
```

```python
# Verificando número ideal de clusters
```

```python
from sklearn.metrics import silhouette_score
```

```python
centr_positions = []
labels = []
inertias = []

for i in range(2,6):
    seeds = np.array(centr.loc[:i-1, :])
    kmean = KMeans(n_clusters=i, init = seeds, n_init = 1, max_iter = 10)
    kmean.fit(data)
    position = kmean.cluster_centers_
    label = kmean.labels_
    inertia = kmean.inertia_
    centr_positions.append(position)
    labels.append(label)
    inertias.append(inertia)
```

```python
centr_positions
```

```python
labels
```

```python
inertias
```

```python
pca.
```

```python
pca = PCA(2)
pc1 = pca.fit_transform(data)[:,0]
pc2 = pca.fit_transform(data)[:,1]
```

```python
fig, axs = plt.subplots(2,2, figsize = (12,8))
axs[0,0].scatter(pc1, pc2, c=labels[0], cmap = 'inferno')
axs[0,0].set_title('k = 2')
axs[0,1].scatter(pc1, pc2, c=labels[1], cmap = 'inferno')
axs[0,1].set_title('k = 3')
axs[1,0].scatter(pc1, pc2, c=labels[2], cmap = 'inferno')
axs[1,0].set_title('k = 4')
axs[1,1].scatter(pc1, pc2, c=labels[3], cmap = 'inferno')
axs[1,1].set_title('k = 5');
```
