## Scores DataFrame

Let's create a DataFrame to store every scores we get by testing some models

```python
def scores(Column, score, precision, recall):
    '''
    Append a new colum of scores into a dataframe of scores, use it to compare your models.
    
    Columns: Name of the columns/model
    score: Accuracy score
    precision: Precision Score
    recall: Recall Score
    '''
    
    df = pd.DataFrame({Column : [score,precision,recall]}, index=['SCORE', 'PRECISION', 'RECALL'])
    x = pd.concat([rain_scores, df], axis = 1)
    return x
```

```python
r = pd.read_csv('./data/rain_data_aus.csv')
```

```python
# adding wind again
w = pd.DataFrame()
for i in range(1,9):
    df_aux = pd.read_csv(f'./data/wind_table_0{i}.csv')
    df_aux.columns = ['date', 'location', 'wind_gustdir', 'wind_gustspeed', 'wind_dir9am',
       'wind_dir3pm', 'wind_speed9am', 'wind_speed3pm']
    w = pd.concat([w,df_aux])
r_w =  pd.merge(r, w, how='left', on=['date', 'location'])
```

```python
# CONVERTING LOCATION INTO LAT|LONG
```

```python
from opencage.geocoder import OpenCageGeocode

key = '46b48d2d3d8f4acb88f582e3a83cbe77'
geocoder = OpenCageGeocode(key)

city_lat = {}
city_lng = {}

for city in r_w['location'].unique():
    query = f'{city}, Australia'
    results = geocoder.geocode(query)
    city_lat[city] =results[0]['geometry']['lat']
    city_lng[city] =results[0]['geometry']['lng']
```

```python
cities = pd.DataFrame(data=[city_lat,city_lng]).T
cities = cities.reset_index()
cities.columns= ['location','lat','lng']
rw = pd.merge(left=r_w,
                    right=cities,
                    on='location')
```

# Applying thresholds to test models

```python
rw = pd.read_csv('./data/rain_data_aus.csv')
rw.drop_duplicates(inplace=True)
rw.drop(columns=['evaporation', 'sunshine', 'cloud9am', 'cloud3pm'],inplace=True)
wind = pd.DataFrame()
for i in range(1,9):
    df_aux = pd.read_csv(f'./data/wind_table_0{i}.csv')
    df_aux.columns = ['date', 'location', 'wind_gustdir', 'wind_gustspeed', 'wind_dir9am',
       'wind_dir3pm', 'wind_speed9am', 'wind_speed3pm']
    wind = pd.concat([wind,df_aux])
rw =  pd.merge(rw, wind, how='left', on=['date', 'location'])
rw.dropna(inplace=True)
import datetime
rw['date']=pd.to_datetime(rw['date'])
```

```python
def funcaozona(row,column):
    dia=((row['date']) - datetime.timedelta(days=1))
    mask = (rw['date']==dia) & (rw['location'] == row['location'])
    result=rw.loc[mask,column]
    if result.shape[0]>0:
        return (rw.loc[mask,column]).values[0]
    else:
        np.nan
```

```python
from tqdm import tqdm
tqdm.pandas()
rw['maxtemp_yesterday']= rw.progress_apply(lambda row: funcaozona(row, 'maxtemp'), axis=1)
```

```python
rw['maxtemp_yesterday'].to_csv('../data/max_temp.csv')
```
