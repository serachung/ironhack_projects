```python
import requests

class BearerAuth(requests.auth.AuthBase):
    def __init__(self, token):
        self.token = token
    def __call__(self, r):
        r.headers["authorization"] = "Bearer " + self.token
        return r
```

## Get Artists

Get an artists snapshot of sample artists based on a given metric. 

Given the metric filter and score range, get a list of artist ids and all relevant metrics for the given range of the respective metric. 

This endpoint is useful for artist profiling based on specific filters such as spotify performance, youtube performance and social media performance.


```python
import time
token_url = !curl -d '{"refreshtoken":"NivvXOOYbKUparpdKQFLdnTXv9d2yRqVEP6fDmILUPxd2yO82QGrUHEiZo75C4kq"}' -H "Content-Type: application/json" -X POST https://api.chartmetric.com/api/token
token = re.findall(r'"([A-Za-z0-9._-]+)"',token_url[-1].split(',')[0])[1]

params_type = 'sp_followers'
df_total=pd.DataFrame()
i = 0

while i < 100000000:
    j = i
    params_list = [f'min={i}','max=100000000','offset=0']
    get_url = f'https://api.chartmetric.com/api/artist/{params_type}/list?{"&".join(params_list)}'
    
    try:
        response = requests.get(get_url, auth=BearerAuth(token))
        df = pd.DataFrame(response.json()['obj']['data'])
        df_total = df_total.append(df).reset_index(drop=True)
    except:
        print('except:', i) 
        token_url = !curl -d '{"refreshtoken":"NivvXOOYbKUparpdKQFLdnTXv9d2yRqVEP6fDmILUPxd2yO82QGrUHEiZo75C4kq"}' -H "Content-Type: application/json" -X POST https://api.chartmetric.com/api/token
        token = re.findall(r'"([A-Za-z0-9._-]+)"',token_url[-1].split(',')[0])[1]
        time.sleep(60)
        response = requests.get(get_url, auth=BearerAuth(token))
        print(response)
        df = pd.DataFrame(response.json()['obj']['data'])
        df_total = df_total.append(df).reset_index(drop=True)
    
    # reassigning i value to the last followers' value 
    i = df_total.loc[df_total.index[-1],'spotify_followers'] + 1

    # saving it as a csv. file
    df_total.to_csv('data/processed/chartmetric_0_5k.csv', index=False)
    
    # break the loop when it reaches the max. value of followers
    if i == j:
        break
    else:
        pass
```

```python
df_total=pd.DataFrame()

for artist_id in artists_df.chartmetric_artist_id:
    get_url = f'https://api.chartmetric.com/api/artist/chartmetric/{artist_id}/get-ids'

    try:
        response = requests.get(get_url, auth=BearerAuth(token))
        df = pd.DataFrame(response.json()['obj'])
        df_total = df_total.append(df).reset_index(drop=True)
    except:
        print('except:', artist_id) 
        
        # refresh token
        token_url = !curl -d '{"refreshtoken":"NivvXOOYbKUparpdKQFLdnTXv9d2yRqVEP6fDmILUPxd2yO82QGrUHEiZo75C4kq"}' -H "Content-Type: application/json" -X POST https://api.chartmetric.com/api/token
        token = re.findall(r'"([A-Za-z0-9._-]+)"',token_url[-1].split(',')[0])[1]
        time.sleep(60)
        
        # request again
        response = requests.get(get_url, auth=BearerAuth(token))
        print(response)
        df = pd.DataFrame(response.json()['obj'])
        df_total = df_total.append(df).reset_index(drop=True)
    
    # saving it as a csv. file
    df_total.to_csv('data/processed/chartmetric_artists_ids.csv', index=False)
```

## Artist - Fan Metrics


Get artist's fan metrics for various services. 

There are two ways to get fan metrics from each source: 

1. Return all fields (by default) 
2. Return only selected field (by adding 'field' parameter) 

For each source, here are the respective fields you can query for: 

SOURCE - VALID FIELDS 

spotify - followers, popularity, listeners 
deezer - fans 
facebook - likes 
twitter - followers 
youtube_channel - subscribers, views, comments, videos (this refers to the artist channel only) 
youtube_artist - views (this includes artist channel and all other videos featuring artist music) 
instagram - followers 
wikipedia - views 
bandsintown - followers 
soundcloud - followers 

If using the field parameter, ensure that the field is valid with respect to the source. 

For example, /wikipedia?field=views is valid, and /wikipedia?field=followers is invalid. 

Note that The API response may not return all the data points between the two dates, when we are missing the data. They may still show on our UI, since we are interpolating the missing data between the two existing data points. See the 'interpolated: true / false' in the response

`id`: chartmetric id

`source`: spotify, deezer, facebook, twitter, instagram, youtube_channel, youtube_artist, wikipedia, bandsintown, soundcloud
