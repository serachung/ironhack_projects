## Get Billboard Artists' Spotify URI

`type`: *Required.* A comma-separated list of item types to search across. 
Valid types are: album , artist, playlist, track, show and episode. 
Search results include hits from all the specified item types. 
For example: q=name:abacab&type=album,track returns both albums and tracks with “abacab” included in their name.

```python
# Finding BILLBOARD artist's uri

artist_list = [artist for artist in bb_artist.artist]

artists=[]
uris = []
followers = []
genres =[]
popularities = []
error_artists = []

for artist in artist_list:
    q = f'artist:{artist}'
    
    try:
        results = sp.search(q=q, limit=1, type='artist')['artists']['items'][0]
    except:
        error_artists.append(artist)
        pass

    artists.append(artist)
    uris.append(results['uri'])
    followers.append(results['followers']['total'])
    genres.append(results['genres'])
    popularities.append(results['popularity'])
```

# Functions and Classes

```python
def get_spotify_uris(df:pd.DataFrame)-> pd.DataFrame:
    '''
    this function receives a dataframe with columns of artists and tracks
    and returns 2 dataframes:
        [0] new df containing spotify unique IDs (uri) for each of the following features:
            - track
            - album
        [1] list with songs that weren't found in spotify
    '''
    artists_col = []    
    tracks_uris = [] 
    track_col = []    
    chart_col = []        
    albums_uris = []
    err_tracks=[]
    
    artist_names = df['artist'].values
    track_names = df['track'].values
    chart_names = df['chart'].values
    
    for i in range(len(artist_names)):
        artist = artist_names[i]
        track = track_names[i]
        q = f'artist:{artist} track: {track}'
        
        try:
            results = sp.search(q=q, limit=1, type='track')
            if results['tracks']['items'][0]['uri'] != '':
                tracks_uris.append(results['tracks']['items'][0]['uri'])
                albums_uris.append(results['tracks']['items'][0]['album']['uri'])
                chart_col.append(chart_names[i])
                artists_col.append(artist)                
                track_col.append(track)
            else:
                chart_col.append(chart_names[i])
                artists_col.append(artist)                
                track_col.append(track)
                err_tracks.append([track, artist, chart_names[i]])
                pass
        except:
            err_tracks.append([track, artist, chart_names[i]])
            pass

    df=pd.DataFrame({'artist': artists_col, 
                'track': track_col, 
                'chart': chart_col, 
                'tracks_uri': tracks_uris, 
                'albums_uris': albums_uris})

    return df, err_tracks
```

```python
def clean_track_title(df:pd.DataFrame)-> pd.DataFrame:
    '''
    This function cleans the song's title column of a dataframe and 
    returns a new dataframe with the right track's title. 
    '''
    
    import re
    
    tracks = []
    
    df = df.reset_index(drop=True)

    for i in range(df.shape[0]):
        string = df.loc[i,'track']
        pattern ='(?: \([A-Za-z0-9 ]+\)|[A-Za-z0-9 :\']* Volume [0-9]+: |, V\.[0-9]+| - [0-9]+\/[0-9]+\/[0-9]+|, \w+)'
        new_track = re.sub(pattern, '', string).strip()
        df.loc[i, 'track'] = new_track
        
    return df
```

```python
def get_audio_features(df:pd.DataFrame)-> pd.DataFrame:
    '''
    This function receives a dataframe containing tracks' Spotify uris 
    and returns a dataframe with the track's audio features and popularity rating.
    '''
    feat_df = pd.DataFrame()
    pop=[]
    
    saved_uris = list(df.tracks_uri)
    
    for uri in saved_uris:
        feat_df = feat_df.append(pd.DataFrame(sp.audio_features(uri)))
        pop.append(sp.track(uri)['popularity'])

    new_df = pd.concat([df, feat_df.reset_index(drop=True), pd.DataFrame(pop, columns=['popularity'])], axis=1)

    return new_df
```

```python
def get_genre_from_album(df: pd.DataFrame)-> pd.DataFrame:
    '''
    This function receives a dataframe containing the 'album_uris' column and
    gets the genres each album was classified, 
    returning the input dataframe with a new 'genres' column.
    '''
    genres = [] 
    for album in df.albums_uris.apply(lambda x : x.split(':')[-1]):
        genres.append(sp.album(album)['genres'])
    
    df['alb_genres'] = genres

    return df
```
